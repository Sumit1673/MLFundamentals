# Linear Regression Models

A linear regression problem with one feature is given as:

$\hat{h} = \theta^{T}X + \theta_{0}$ this is called as hypothesis function. This equation tells what our model thinks of our input data and teach it about wrong and right we need function which tells how wrong or right you are about the value you calculated.

Thats given by the loss function:
$L = (\hat{h} - y)^2$

Calculating loss for each sample comes under SGD based learning. where each sample is passed to the network, hypothesis is calculated, loss is found and $\theta^{T}$ is manipulated.

Another type of learning where a batch of samples are passed to the equation and then loss is calculated. And based on the combined loss $\theta^{T}$ is manipulated. So, to calculate that loss we need below equation.

If m samples are present, then it is known as cost funtion and is calculated as:

$J(X, h_{\theta}) = \frac{1}{m}\sum_{i=1}^{m}(\theta^{T}x^{(i)} - y^{(i)})^2$

So our ultimate goal is to find that $\theta$ which minimizes the cost function.

To find $\theta$ there is a closed form solution called as Normal Equation.

$\hat{\theta} = (X^TX)^{-1}X^{T}y$

<img src="normal_equation.png">


**INPUT VECTOR** is always added with a column vector of ones as a bias

```python
X_b = np.c_[np.ones((100, 1)), X] # add x0 = 1 to each instance. Its just like bias
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

**function uses np.linalg.lstsq to find the theta value. lstsq="least squares"**

lstsq internally uses pseudo inverse method to find best theta value which can
be called by np.linalg.pinv()

SO, np.linalg.lstsq()--> np.linalg.pinv() --> numpy.linalg.svd() 

The pseudoinverse itself is computed using a standard matrix factorization technique called Singular Value Decomposition (SVD) that can decompose the training set matrix X into the matrix multiplication of three matrices U Σ V T (see numpy.linalg.svd() ). The pseudoinverse is computed as $X + = V\sum_ + UT$ . To compute the matrix Σ + , the algorithm takes Σ and sets to zero all values smaller than a tiny
threshold value, then it replaces all the non-zero values with their inverse, and finally it transposes the resulting matrix. This approach is more efficient than computing the Normal Equation, plus it handles edge cases nicely: indeed, the Normal Equation may not work if the matrix $X^TX$ is not invertible (i.e., singular), such as if m < n or if some features are redundant, but the pseudoinverse is always defined.

# 1. Gradient Descent

Iterative optimization algorithm to minimize cost function.

Gradient --> gives the direction to the algorithm +/- where to descent. And 
how long step to be taken in that direction is determined my learning rate.

Small LR --> Long time.           High LR --> Oscillations around the minima, model diverge

When $\theta$ reaches minima --> model has converged.

$ \theta := \theta - \alpha(\frac{\partial{J(\theta, b}}{\partial{\theta}})$

**Complexity with GD**

<img src='GD1.png'>

If weights are randomly initialized and if it starts from left side local minima is found which could be wrong
as we are looking for global minima. If weights starts from right GD will take longer time to come out of that pleatue and
find global minima.

If a cost function is convex --> No local minima only global minima. (MSE for LinReg)

The cost function decides the shape. If the features are not scaled the cost function will elongted, GD may hit plateau
but not in the case where cost function is convex (all features are of same scale) i.e. no matter where you initialize the weights you will reach a close value to the global minima.

Figure shows the model parameter space. Depends upon number of parameters.

Number of Feautes high --> Model Params (high) --> Longer and harder to find the global minima


# 1.1 BATCH GD
Definition:
    
update weights =subtrat the current weights from the derivate wrt all weights(only when complete dataset is passed to the network and the loss is calculated.)

At each step complete data is forward passed, loss is calculated and then BGD is performed. It is very slow.

Formula:
<img src='bgd.png'>



```python
for lr_i in [0.1, 0.01, 0.001]:
    theta = np.random.randn(X_b.shape[1], 1)
    ite = 100
    n_samples = 100
    plt.plot(X, y, 'b.', label="data")
    x_test = np.array([[0],[2]]) # 2 test sample
    # add bias
    x_test_b = np.c_[np.ones((2,1)), x_test]
    for i in range(ite):
        # in BGD at every step full data is passed to the n/w loss is calculate
        # and wights are updates. And to minimize the cost function we need global minima 
        # and to find that we use here normal equation since number of samples are less
        # using the above formula
        grads = 2/n_samples*X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - lr_i*grads
        if i%4==0:
            prediction = x_test_b.dot(theta)
            plt.plot(x_test, prediction, 'g*')
            plt.plot(x_test, prediction, 'r-')
#     theta
    plt.show()
```

# SGD
<img src='sgd.png'>

Take one example pass it to the network compute loss, calculate the grdients and update the weights.
So every iter we make progress. Unlike BGD where we wait for all the examples to pass to the network and at the end loss wrt to each 
examples are computed, summed, calculate gradients and at the end update the weights.

Path to the global minima is zig zag and sometimes it keeps on oscillating near the global minima but thats not the problem.

```python
t0, t1 = 5, 50
# learning schedule hyperparameters
def learning_schedule(t):
    return t0 / (t + t1)
lr = 0.01
n_epochs = 50
etheta = []
for ep in range(n_epochs):
    for i in range(n_samples):
        random_sample_i = np.random.randint(n_samples)
        xi = X_b[random_sample_i:random_sample_i+1]
        yi = y[random_sample_i:random_sample_i+1]

        grads = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(ep * n_samples + i)
        theta = theta - lr * grads
        etheta.append(theta)
theta
```

# MiniBGD

at each step, instead of computing the gradients based on the full train‐
ing set (as in Batch GD) or based on just one instance (as in Stochastic GD), Mini-
batch GD computes the gradients on small random sets of instances called mini-
batches. The main advantage of Mini-batch GD over Stochastic GD is that you can
get a performance boost from hardware optimization of matrix operations, especially
when using GPUs.


# Polynomial Regression
For complex data,

Surprisingly, you can actually use a linear model to fit nonlinear data. A simple way to do this is to
add powers of each feature as new features, then train a linear model on this extended
set of features. This technique is called Polynomial Regression.

# **What degree model to choose when the origin of data is unknown** --> Learning Curves

In the above experiment the data was generated using 2 deg polynomial but to fit the
model we used 300 deg polynomial. So we over-trained the model in this case and it
overfitted. 

**But generally we never know what generated the data.**

1. One way using the cross validation technique.
2. LEARNING CURVES: Plots of the model performance during training on different size of the training dataset.



```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
    
    train_errors, val_errors = [], []
    # Everytime the model is train on different dataset and is validated on the
    # same validation set
    for m in range(1, len(X_train)):
        x_i, y_i = X_train[:m], y_train[:m]
        # train the model
        model.fit(x_i, y_i)
        # predict on train data
        train_pred = model.predict(x_i)
        
        val_pred = model.predict(X_val)
        
        train_errors.append(mean_squared_error(y_i, train_pred))
        val_errors.append(mean_squared_error(y_val, val_pred))
        
    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="Train")
    plt.plot(np.sqrt(val_errors), "b-+", linewidth=3, label="Val")
    plt.legend(loc="upper right")
    
```

```python
model = LinearRegression()
plot_learning_curves(model, X, y)
plt.axis([0, 80, 0, 3])
```

```python
t1a, t1b, t2a, t2b = -1, 3, -1.5, 1.5

t1s = np.linspace(t1a, t1b, 500)
t2s = np.linspace(t2a, t2b, 500)
t1, t2 = np.meshgrid(t1s, t2s)
T = np.c_[t1.ravel(), t2.ravel()]
Xr = np.array([[1, 1], [1, -1], [1, 0.5]])
yr = 2 * Xr[:, :1] + 0.5 * Xr[:, 1:]

J = (1/len(Xr) * np.sum((T.dot(Xr.T) - yr.T)**2, axis=1)).reshape(t1.shape)

N1 = np.linalg.norm(T, ord=1, axis=1).reshape(t1.shape)
N2 = np.linalg.norm(T, ord=2, axis=1).reshape(t1.shape)

t_min_idx = np.unravel_index(np.argmin(J), J.shape)
t1_min, t2_min = t1[t_min_idx], t2[t_min_idx]

t_init = np.array([[0.25], [-1]])
```

```python
def bgd_path(theta, X, y, l1, l2, core = 1, eta = 0.05, n_iterations = 200):
    path = [theta]
    for iteration in range(n_iterations):
        gradients = core * 2/len(X) * X.T.dot(X.dot(theta) - y) + l1 * np.sign(theta) + l2 * theta
        theta = theta - eta * gradients
        path.append(theta)
    return np.array(path)

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(10.1, 8))
for i, N, l1, l2, title in ((0, N1, 2., 0, "Lasso"), (1, N2, 0,  2., "Ridge")):
    JR = J + l1 * N1 + l2 * 0.5 * N2**2
    
    tr_min_idx = np.unravel_index(np.argmin(JR), JR.shape)
    t1r_min, t2r_min = t1[tr_min_idx], t2[tr_min_idx]

    levelsJ=(np.exp(np.linspace(0, 1, 20)) - 1) * (np.max(J) - np.min(J)) + np.min(J)
    levelsJR=(np.exp(np.linspace(0, 1, 20)) - 1) * (np.max(JR) - np.min(JR)) + np.min(JR)
    levelsN=np.linspace(0, np.max(N), 10)
    
    path_J = bgd_path(t_init, Xr, yr, l1=0, l2=0)
    path_JR = bgd_path(t_init, Xr, yr, l1, l2)
    path_N = bgd_path(np.array([[2.0], [0.5]]), Xr, yr, np.sign(l1)/3, np.sign(l2), core=0)

    ax = axes[i, 0]
    ax.grid(True)
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    ax.contourf(t1, t2, N / 2., levels=levelsN)
    ax.plot(path_N[:, 0], path_N[:, 1], "y--")
    ax.plot(0, 0, "ys")
    ax.plot(t1_min, t2_min, "ys")
    ax.set_title(r"$\ell_{}$ penalty".format(i + 1), fontsize=16)
    ax.axis([t1a, t1b, t2a, t2b])
    if i == 1:
        ax.set_xlabel(r"$\theta_1$", fontsize=16)
    ax.set_ylabel(r"$\theta_2$", fontsize=16, rotation=0)

    ax = axes[i, 1]
    ax.grid(True)
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    ax.contourf(t1, t2, JR, levels=levelsJR, alpha=0.9)
    ax.plot(path_JR[:, 0], path_JR[:, 1], "w-o")
    ax.plot(path_N[:, 0], path_N[:, 1], "y--")
    ax.plot(0, 0, "ys")
    ax.plot(t1_min, t2_min, "ys")
    ax.plot(t1r_min, t2r_min, "rs")
    ax.set_title(title, fontsize=16)
    ax.axis([t1a, t1b, t2a, t2b])
    if i == 1:
        ax.set_xlabel(r"$\theta_1$", fontsize=16)

# save_fig("lasso_vs_ridge_plot")
plt.show()
```

# Logistic(Logit) Regression

Difference between Linear regression and logistic regression is instead of producing a direct value 
as in linear regression, logistic regression produces a probability for an instance which classifies
whether an instance belongs to a particular class or not. 

To produce the output the logistic regression uses a non-linear functions (like sigmoid, relu etc.)
bounded by a threshold value.

Unlike Linear regression where we have a closed form equation called as Normal Equation which helps in 
minimizing the cost function by finding an optimum $\theta$ value, in logistic regression we dont have 
such equation.

Unlike lineReg where the cost function may or may not be convex, here the cost function is convex,
so gradient descent and other optimization algos is gauranteed to find global minima.

**Probability Estimation**

$\hat{p} = h_{\theta}(x) = \sigma(x^T\theta)$

**Sigmoid Function**

$\sigma(t) = \frac{1}{1 + exp(-t)}$

**Model Prediction**

y = 0  $ if \hat{p} < 0.5 $
y = 1 $ if \hat{p} > 0.5$

**Logistic Regression cost function (log loss)**

$J(\theta) = -\frac{1}{m}\sum_{i=1}^{m}[y^(i)log(\hat{p}^i) + (1-y^i)log(1-\hat{p}^i)]$



```python
from sklearn import datasets
iris = datasets.load_iris() # 3 classes
#np.unique(iris['target'])
#np.unique(iris['target_names'])
#np.unique(iris['DESCR'])\
# iris["data"]
X = iris["data"][:, 3:] # petal width
y = (iris["target"] == 2).astype(np.int) # 1 if Iris-Virginica, else 0
print(iris.DESCR)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X, y)

X_new = np.linspace(0, 3, 1000).reshape(-1, 1) # converts (1000,) --> (1000, 1)
y_proba = log_reg.predict_proba(X_new)
# print(y_proba[:,1])
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")

# + more Matplotlib code to make the image look pretty
```

```python
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0]
plt.figure(figsize=(8, 3))
plt.plot(X[y==0], y[y==0], "bs")
plt.plot(X[y==1], y[y==1], "g^")
plt.plot([decision_boundary, decision_boundary], [-1, 2], 'k:', linewidth=2)
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")
plt.text(decision_boundary+0.02, 0.15, "Decision  boundary", fontsize=14, color="k", ha="center")
plt.arrow(decision_boundary, 0.08, -0.3, 0, head_width=0.05, head_length=0.1, fc='b', ec='b')
plt.arrow(decision_boundary, 0.92, 0.3, 0, head_width=0.05, head_length=0.1, fc='g', ec='g')
plt.xlabel("Petal width (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 3, -0.02, 1.02])
# save_fig("logistic_regression_plot")
plt.show()
```
