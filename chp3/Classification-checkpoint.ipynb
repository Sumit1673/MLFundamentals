# Performance Measurement

 **Measuring Accuracy Using Cross-Validation**
 Creating more controlled Cross-Validation using StratifiedKFold

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

stratified_split = StratifiedKFold(n_splits=3, random_state=42)

for train_indx, test_indx in stratified_split.split(X_train, y_train_5):
    clone_clf = clone(sgd_clf)
    train_X_sf = X_train[train_indx]
    train_y_sf = y_train_5[train_indx]
    test_X_sf = X_train[test_indx]
    test_y_sf = y_train_5[test_indx]
    
    clone_clf.fit(train_X_sf, train_y_sf)
    predictions = clone_clf.predict(test_X_sf)
    correct = sum(predictions == test_y_sf)
    
    print('Acc: {}'.format(correct/len(test_y_sf)))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring='accuracy')
```

```python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
```

**cross_val_pred**
cross_val_predict creates K fold training set, runs the classifier on each set 
and return the prediction from each set of K folds. 

What does cross_val_predict return?
It uses KFold to split the data to k parts and then for i=1..k iterations:

takes i'th part as the test data and all other parts as training data
trains the model with training data (all parts except i'th)
then by using this trained model, predicts labels for i'th part (test data)
In each iteration, label of i'th part of data gets predicted. In the end cross_val_predict merges all partially predicted labels and returns them as the final result.

# Metrics for better performance measurement based on elements of Confusion Matrix

# Precision

**Precision quantifies the number of positive class predictions that actually belong to the positive class**

$ Precision = \frac{TP}{TP + FP}$

$Precision = \frac{true positive}{no. of predicted positive}$

# Recall (Sensitivity or True Positive Rate)

**This metric corresponds to the proportion of positive data points that are correctly considered as positive, with respect to all positive data points. In other words, the higher TPR, the fewer positive data points we will miss.**


$ Recall = \frac{TP}{TP + FN}$

$ Recall = \frac{true positive}{no. of actual positive}$

# F1 Score

**Is a way of using both precision and recall together. This could be good when compairing classifiers.**

$F1  = \frac{1}{\frac{1}{precision} + \frac{1}{recall}}$

# False Positive Rate (FPR = 1 - True Negative Rate (TNR / Specificity) also called as Fallout

**Intuitively this metric corresponds to the proportion of negative data points that are mistakenly considered as positive, with respect to all negative data points. In other words, the higher FPR, the more negative data points will be missclassified.**

$FPR  = \frac{FP}{FP+TN}$

```python
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3,method="decision_function")
```

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
    plt.legend(loc="center right", fontsize=8)
    plt.grid(True)
    plt.xlabel("Threshold")
    plt.ylabel("P and R")
    plt.axis([-50000, 50000, 0, 1]) 

plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
plt.show()
```

```python
# np.argmax gives the index of first recall which has precision>0.90
recall_with_90_precision = recalls[np.argmax(precisions >= 0.90)]
print('Recall with 0.90 Precision {}'.format(recall_with_90_precision))
# np.argmax gives the index of first threshold which has precision>0.90
threshold_with_90_precision = thresholds[np.argmax(precisions >= 0.90)]
print('Threshold with 0.90 Precision {}'.format(threshold_with_90_precision))

plot_precision_recall_vs_threshold(precisions, recalls, thresholds)

plt.plot([threshold_90_precision, threshold_90_precision], [0., 0.9], "r:")  
# # Not shown
plt.plot([-50000, threshold_with_90_precision], [0.9, 0.9], "g:")     
# # Not shown
plt.plot([-50000, threshold_with_90_precision],
         [recall_with_90_precision, recall_with_90_precision], "r:")# Not shown


plt.plot([threshold_with_90_precision], [0.9], "ro")  
plt.annotate("Th with 0.9 Pre", (threshold_with_90_precision, 0.9))
# # Not shown
plt.plot([threshold_with_90_precision], [recall_with_90_precision], "bo") 
plt.annotate("Rec with 0.9 Pre", (threshold_with_90_precision, recall_with_90_precision))
# Not shown
# save_fig("precision_recall_vs_threshold_plot")                                              # Not shown
plt.show()
```

Using this curve we can decide what we need. If we need 90% of both recall and precision we need threshold with a value at the red point. We can clearly find out what we need by plotting recall vs precision

```python
y_train_pred_90 = (y_scores >= threshold_with_90_precision)
```

# Receiver Operating Characteristics (ROC) Curve

1. Used with binary classifier

2. A plot between Recall (True positive rate, TPR) and False positive rate (FPR). FPR = ratio of negative instances that are incorrectly classified as positive. It is equal to one minus the true negative rate, which is the ratio of negative instances that are correctly classified as negative. The TNR is also called specificity. Hence the ROC curve plots sensitivity (recall) versus 1  specificity.
3. ROC summaries all confusion matrix for different thresholds.

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
print(fpr, '\n', tpr)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    
    #some random classifier
    plt.plot([0, 1], [0, 1], 'k--') # dashed diagonal # line between min and the max value
    plt.axis([0, 1, 0, 1])    
    plt.xlabel('False Positive Rate (Fall-Out)', fontsize=16) # Not shown
    plt.ylabel('True Positive Rate (Recall)', fontsize=16)    # Not shown
    plt.grid(True)                                            # Not shown

plt.figure(figsize=(8, 6))                                    # Not shown
plot_roc_curve(fpr, tpr)

fpr_90 = fpr[np.argmax(tpr >= recall_90_precision)]           # Not shown
plt.plot([fpr_90, fpr_90], [0., recall_90_precision], "r:")   # Not shown
plt.plot([0.0, fpr_90], [recall_90_precision, recall_90_precision], "r:")  # Not shown
plt.plot([fpr_90], [recall_90_precision], "ro")               # Not shown
# save_fig("roc_curve_plot")                                    # Not shown
plt.show()
```

1. Higher the recall --> higher the false positive rate (neg instances incorrectly classified as positive)

2. a good classifier will be away from this diagonal line.

3. The Area Under the Curve (AUC) is the measure of the ability of a classifier to distinguish between classes and is used as a summary of the ROC curve. The higher the AUC, the better the performance of the model at distinguishing between the positive and negative classes.
4. AUC stands for "Area under the ROC Curve." That is, AUC measures the entire two-dimensional area underneath the entire ROC curve (think integral calculus) from (0,0) to (1,1).
<img src='https://developers.google.com/machine-learning/crash-course/images/AUC.svg'>




The above curve is ROC and area under it is AUC.

# MULTI-CLASS CLASSIFICATION

**One-Versus-All (OvA):**

Train a binary classifier for each class. Prediction will be based on the highest score produced by the classifiers.

**One-Versus-One (OvO):**

Train a binary classifier for every pair of classes. For example: 0 and 1, 0 and 2 etc..

**SciKit Learn**

Predicts itself whethter the problem is binary or of multi-class and runs internally **OvA** (excepts SVM where it uses OvO).

```python
sgd_clf.fit(X_train, y_train) # y_train, not y_train_5
sgd_clf.predict([some_digit])
```
